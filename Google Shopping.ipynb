```python
####### Use Selenium to open webpages that use java script rendering ###############
from selenium import webdriver
import requests
import time
from urllib.request import Request, urlopen
from selenium.webdriver.common.keys import Keys
####### Use Pandas to manage and arrange data #################
import pandas as pd
####### Use BeautifulSoup to read and navigate the DOM ################
from bs4 import BeautifulSoup
from bs4.element import Comment
import urllib.request
import urllib.error
######### Use Regex to clean up some of the elements we scrape #################
import re
############# Use Matplotlib to create and display a price graphs ################
import matplotlib.pyplot as plt
######### Use counter to count results of vendors ########
from collections import Counter
```

```python
########## First ask user to input term to search Google Shopping ##########
results = []
pages = []
allsellers = []
terms = input('Enter serach term(s) here: ')
terms = terms.split(', ')
########## For each comma sepearated term retrive first page results ##############
for t in terms:
    driver = webdriver.Chrome()
    base = 'https://www.google.com/search?q='
    shop = '&tbm=shop&tbs=vw:l'
    url = base + t + shop
    driver.implicitly_wait(30)
    driver.get(url)
    soup_level1=BeautifulSoup(driver.page_source, 'lxml')



    data = soup_level1.find_all('div', {'class':'mQ35Be'})
    data1=[re.sub(r'<.+?>',r'',str(a)) for a in data]
    data1=[re.sub(r'.*(stores).*',r'',str(a))for a in data1]
    data1=[re.sub(r'.*(positive).*',r'',str(a))for a in data1]
    data1=[x for x in data1 if x]
    nearby='Also available nearby'
    caturl=[re.sub(r' ',r'\%20',str(a))for a in url]
    caturl=''.join(caturl)
    vendors = [re.sub(r'.*from',r'',str(a))for a in data1]
    print('Search url : ' + caturl)
    while nearby in data1:
        data1.remove(nearby)
    print(data1)
    print(len(data1))

    products=soup_level1.find_all('div', {'class':'eIuuYe'})
    products=[re.sub(r'href="',r'\^https://www.google.com',str(a)) for a in products]
    products=[re.sub(r'.*\^',r'',str(a)) for a in products]
    products=[re.sub(r'".*',r'',str(a)) for a in products]
    products=[re.sub(r'.*(aclk).*',r'',str(a)) for a in products]
    products=[re.sub(r'\?.*',r'',str(a))for a in products]
    products=[x for x in products if x]

    results.append(data1)
    pages.append(products)

    driver.quit()
############## For single listing sellers the Vendor name and price is returned. Grouped products are stored in a list of lists for wach search term ################
print(len(results))
print(len(pages))
print(len(pages[0]))
```

```python
########## Select the first set of grouped vendors from the search results ##########
########## for each listing the vendors are counted and Price, GTIN and Barcode are gathered ########
for i in pages[0]:
    url = i
    page = urllib.request.urlopen(url)
    soup = BeautifulSoup(page, 'html.parser')
    title= soup.find('h1', {'id':'product-name'})
    title=[re.sub(r'<.+?>',r'',str(a)) for a in title]
    print(title)
    table = soup.find_all('td', {'class':'os-seller-name'})
    y=[re.sub(r'<.+?>',r'',str(a)) for a in table]
    seller=len(y)
    seller=str(seller)
    print('Sellers : ' + seller)
    print(url)
    ######################### Rankings and Chart ######################

    chart=pd.read_html(url)
    chart=chart[2]
    header=chart.iloc[0]
    chart=chart[1:]

    chart.columns = header

    chart[['Total Price']]=chart[['Total Price']].replace('[\$,]', '', regex=True).astype(float)
    Sellers=chart.iloc[:,0:1]
    Sellers=pd.DataFrame(Sellers)
    Sellers=Sellers['Sellers'].tolist()
    vendors.extend(Sellers)

    Price=chart.iloc[:,4:]
    Price=pd.DataFrame(Price)
    Price=Price['Total Price'].tolist()
    avg = chart['Total Price'].sum()
    avgs = float(seller)
    avg = float(avg)
    avg = (avg/avgs)

#     print(Price)
    ##################### Create Price Visualization for PLA ##############################
    ax = chart[['Sellers','Total Price']].plot(kind='barh', legend=True)

    ax.set_ylabel('Seller')
    ax.set_yticklabels(Sellers)
    ax.set_xlabel('Cost')
    ax.invert_yaxis()
    for i, v in enumerate(Price):
        ax.text(v + 3, i + .25, str(v), color='blue', fontweight='bold')

    plt.show()
    ##################### Look For Spcifications in Listings ##############################

    specs= url + '/specs'
    table = pd.read_html(specs)
    table=table[-1]
    try:
        mpn=table.loc[1,1]
    except KeyError:
        mpn='--'

    try:
        GTIN=table.loc[1,1]
        GTIN=GTIN.lstrip('0')
    except KeyError:
        GTIN='--'
    print('MPN : '+mpn+'   GTIN : '+GTIN)
    print('')
########################## Create Pie Graph of Seller share on first page reuslts ##################
print('PLA Share on First Page results for search term : '+terms[0])
count=Counter(vendors)
count=dict(count.most_common(10))
# print(count.keys())
labels=count.keys()
sizes=count.values()
plt.pie(sizes, labels=labels, shadow=True, autopct='%1.1f%%', startangle=45)
plt.axis('equal')
plt.show
```
